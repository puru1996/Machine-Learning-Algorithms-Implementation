```python
import numpy as np 
import pandas as pd 
import math
from matplotlib import pyplot as plt 
```

<h2> Creating the code for Multi-Layer Perceptron </h2>

```python
np.random.seed(0)

def tanh_function(x):
    return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))

def der_tanh(a):
    return 1 - tanh_function(a) * tanh_function(a)

def sigmoid_func(x):
    return 1/(1+np.exp(-x))

def der_sigmoid(x):
    return x*(1-x)

def costfunc(prediction, trueval):
    return trueval- prediction


def multilayer_perceptron (inp, out, hid, epochs, learning_rate , act_func, der_act):
    X= inp
    Y = out.T
    
    #Shape of the weight vector defined
    dim_1, dim_2 = X.shape
    
    #Defining the number of Hidden units 
    hidden_units = hid
    
    #Defining the Output vector Dimension
    output_dimension = len(Y.T)
    
    #Initializing the weight between the input vector and hidden units as random
    W1 = np.random.random((dim_2, hidden_units))
    
    #Initializing the weight between the hidden units and the output vector as random 
    W2 = np.random.random((hidden_units, output_dimension))
    

    #FORWARD PROPAGATION
    for i in range(epochs):
        input_layer  = X
        input_hidden = act_func(np.dot(input_layer, W1))
        hidden_output = act_func(np.dot(input_hidden, W2))

        #BACKWARD PROPAGATION

        #Computing the Error between the final output and the Actual value
        hidden_output_error = costfunc(hidden_output, Y)
        hidden_output_delta = hidden_output_error*der_act(hidden_output)
        #Computing the delta between the input and the hidden layer 
        input_hidden_error = np.dot(hidden_output_delta, W2.T)
        input_hidden_delta = input_hidden_error * der_act(input_hidden)

        #Weight Updation 
        W2 = W2 + learning_rate *np.dot(input_hidden.T, hidden_output_delta)
        W1 = W1 + learning_rate *np.dot(input_layer.T, input_hidden_delta )
        
        if i>0 and i%2000==0:
            print (i, 'Epochs Completed')
        
    return W1, W2
```

```python
def multilayer_perceptron_mod_xor (inp, out, hid, epochs, learning_rate , test_input , test_output, act_func, der_act):

    X= inp
    Y = out.T
    
    #Shape of the weight vector defined
    dim_1, dim_2 = X.shape
    
    #Defining the number of Hidden units 
    hidden_units = hid
    
    #Defining the Output vector Dimension
    output_dimension = len(Y.T)
    
    #Initializing the weight between the input vector and hidden units as random
    W1 = np.random.random((dim_2, hidden_units))
    
    #Initializing the weight between the hidden units and the output vector as random 
    W2 = np.random.random((hidden_units, output_dimension))
    
    f= {}
    df_accuracy = pd.DataFrame(columns = ['Accuracy'])
    df_weights = pd.DataFrame(columns = ['W1', 'W2', 'W3' , 'W4', 'W5'])
    df_output = pd.DataFrame(columns = ['Bit1', 'Bit2', 'Bit3', 'Bit4'])
    #FORWARD PROPAGATION
    for i in range(epochs):
        input_layer  = X
        input_hidden = act_func(np.dot(input_layer, W1))
        hidden_output = act_func(np.dot(input_hidden, W2))

        #BACKWARD PROPAGATION

        #Computing the Error between the final output and the Actual value
        hidden_output_error = costfunc(hidden_output, Y)
        hidden_output_delta = hidden_output_error*der_act(hidden_output)

        input_hidden_error = np.dot(hidden_output_delta, W2.T)
        input_hidden_delta = input_hidden_error * der_act(input_hidden)

        #Weight Updation 
        W2 = W2 + learning_rate *np.dot(input_hidden.T, hidden_output_delta)
        W1 = W1 + learning_rate *np.dot(input_layer.T, input_hidden_delta )
        df_weights.loc[i] = [W2[0][0], W2[1][0], W2[2][0], W2[3][0], W2[4][0]]
        
        count = 0
        output = []
        for x,y in zip(test_input,test_output.T):
            input_hidden_prediction = act_func(np.dot(W1.T, x))
            final_prediction = act_func(np.dot(W2.T, input_hidden_prediction))
            output.append(final_prediction)
            if int(final_prediction>0.5) == y:
                count = count+1
        df_accuracy.loc[i] = [count/len(test_input)*100]
        df_output.loc[i] = [output[0][0], output[1][0], output[2][0] , output[3][0] ]
         
        
    return df_weights, df_accuracy, df_output
```

<h2> Modifying my MLP code so that it stores the weight at each iteration so that i can see how the weight update and change in error with each epoch occurs </h2>

```python
def multilayer_perceptron_mod (inp, out, hid, epochs, learning_rate , test_input , test_output, act_func, der_act):

    X= inp
    Y = out.T
    
    #Shape of the weight vector defined
    dim_1, dim_2 = X.shape
    
    #Defining the number of Hidden units 
    hidden_units = hid
    
    #Defining the Output vector Dimension
    output_dimension = len(Y.T)
    
    #Initializing the weight between the input vector and hidden units as random
    W1 = np.random.random((dim_2, hidden_units))
    
    #Initializing the weight between the hidden units and the output vector as random 
    W2 = np.random.random((hidden_units, output_dimension))
    
    f= {}
    df_accuracy = pd.DataFrame(columns = ['RMSE', 'R2'])
    df_weights = pd.DataFrame(columns = ['W1', 'W2', 'W3' , 'W4', 'W5'])
    #FORWARD PROPAGATION
    for i in range(epochs):
        input_layer  = X
        input_hidden = act_func(np.dot(input_layer, W1))
        hidden_output = act_func(np.dot(input_hidden, W2))

        #BACKWARD PROPAGATION

        #Computing the Error between the final output and the Actual value
        hidden_output_error = costfunc(hidden_output, Y)
        hidden_output_delta = hidden_output_error*der_act(hidden_output)

        input_hidden_error = np.dot(hidden_output_delta, W2.T)
        input_hidden_delta = input_hidden_error * der_act(input_hidden)

        #Weight Updation 
        W2 = W2 + learning_rate *np.dot(input_hidden.T, hidden_output_delta)
        W1 = W1 + learning_rate *np.dot(input_layer.T, input_hidden_delta )
        df_weights.loc[i] = [W2[0][0], W2[1][0], W2[2][0], W2[3][0], W2[4][0]]
        predictions = []
        actual_vals = []
        for x,y in zip(test_input,test_output.T):
            input_hidden_prediction = act_func(np.dot(W1.T, x))
            final_prediction = act_func(np.dot(W2.T, input_hidden_prediction))
            predictions.append(final_prediction)
            actual_vals.append(y) 
        from sklearn.metrics import mean_squared_error
        RMSE = math.sqrt(mean_squared_error(actual_vals, predictions))
        from sklearn.metrics import r2_score
        R2_Score = r2_score(actual_vals, predictions)
        df_accuracy.loc[i] = [RMSE, R2_Score]

    return df_weights, df_accuracy
```
